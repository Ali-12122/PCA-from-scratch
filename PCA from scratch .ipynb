```python
X_meaned = X - np.mean(X , axis = 0)
```

```python
cov_mat = np.cov(X_meaned , rowvar = False)   # rowvar = False >> work with X_meaned as transposed
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

```python
X_meaned.shape
```

```python
#sort the eigenvalues and eigenvectors in descending order
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]
sorted_eigenvectors = eigen_vectors[:,sorted_index]
```

```python
n_components = 2 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:, 0:n_components]
```

```python
eigenvector_subset.transpose().shape
```

```python
X_reduced = np.dot(eigenvector_subset.transpose(), X_meaned.transpose()).transpose()
```

```python
X_reduced.shape
```

```python
sum_cov_mat = np.zeros((np.shape(eigen_vectors)[0], np.shape(eigen_vectors)[1] + 1))
for i in range(np.shape(eigen_vectors)[0]):
    X_projected = np.dot(eigen_vectors[i], X_meaned.transpose()) 
    X_projected_cov = np.cov(X_projected , rowvar = False)
    sum_cov_mat[i, 0] = np.sum(X_projected_cov)
    sum_cov_mat[i, 1:] = eigen_vectors[i]
sum_cov_mat
```

```python
sum_cov_mat = np.sort(sum_cov_mat, axis=0)
sum_cov_mat
```
